```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# Path to one sample MRI image from the dataset
image_path = r"C:\Users\ASUS\Brain Tumor ML Model\yes\Y1.jpg"

if not os.path.exists(image_path):
    raise FileNotFoundError(f"❌ File not found: {image_path}")
```

```python
# Load image
img = cv2.imread(image_path)
if img is None:
    raise ValueError(f"❌ Failed to load image at {image_path}")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Step 2: Reduce noise
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
# Step 3: Threshold to detect bright regions
_, thresh = cv2.threshold(blur, 45, 255, cv2.THRESH_BINARY)
```

```python
# Step 4: Morphological operations to remove small noise
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
# Step 5: Find contours
contours, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

if contours:
    # Get the largest contour (assuming tumor is largest white blob)
    c = max(contours, key=cv2.contourArea)

    # Get bounding box coordinates
    x, y, w, h = cv2.boundingRect(c)

    # Draw rectangle on the original image
    output = img.copy()
    cv2.rectangle(output, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Step 6: Show results
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original MRI")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Detected Bounding Box")
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
else:
    print("No tumor detected in this image.")
```
